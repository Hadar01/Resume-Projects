```python
swiggy = swiggy[['ID', 'Area' , 'City' , 'Restaurant', 'Price', 'Avg_ratings', 'Food_type', 'Address']]
```

```python
swiggy.head(3)
```

```python
def convert3(obj):
    l=[]
    l.append(obj)
    return l
```

```python
swiggy['Area'] = swiggy['Area'].apply(convert3)
```

```python
swiggy['City'] = swiggy['City'].apply(convert3)
swiggy['Rest'] = swiggy['Restaurant'].apply(convert3)
swiggy['Price'] = swiggy['Price'].apply(convert3)
swiggy['Avg_ratings'] = swiggy['Avg_ratings'].apply(convert3)
swiggy['Food'] = swiggy['Food_type'].apply(convert3)
swiggy['Address'] = swiggy['Address'].apply(convert3)
```

```python
swiggy.head(3)
```

```python
swiggy['Rest'] = swiggy['Rest'].apply(lambda x:[i.replace(" ", "") for i in x])
swiggy['Address'] = swiggy['Address'].apply(lambda x:[i.replace(" ", "") for i in x])
swiggy['Food'] = swiggy['Food'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
swiggy.head(3)
```

```python
swiggy['tags'] = swiggy['Food']+ swiggy['Area']+swiggy['Address']+swiggy['Rest']+swiggy['City']
```

```python
swiggy.head(3)
```

```python
new_df = swiggy[['ID', 'Food_type', 'Restaurant', 'tags']]
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(3)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head(3)
```

```python
new_df.tail(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 8680, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(food):
    food_index = new_df[new_df['Food_type'] == food].index[0]
    distances = similarity[food_index]
    food_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    print("Top 5 foods and there restaurants are:-\n")
    for i in food_list:
        print(f"Food:{new_df.iloc[i[0]].Food_type} -> restraunt: {new_df.iloc[i[0]].Restaurant} ")
```

```python
recommend('Chinese,Snacks,Tandoor')
```
